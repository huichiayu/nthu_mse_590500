## Machine leanring algorithms

### Machine learning is a field of study that gives computers the ability to learn from data without explicitly programmed. – Arthur Sameul (1959)

-- A numerical model that gradually adjusts parameters accroding to the correctness of the results.


**Key characters:**
- Use data.
- Improve over time with more examples.
- Generalize to unseen data.
- Enable prediction.
- Enable generation.


**Types of ML:**
- Supervised learning
    - Provide the learning method with training data and appropriate labels to fit
    - Classification: model fits training cases to the correct classes/labels
    - Regression: model fits training cases to target variable
- Unsupervised learning
    - Provide learning method with data and no known labels, the goal is to find patterns or relationships in the data
    - Clustering: find groups within the data in some N-dimensional space
    - Dimensionality reduction: PCA, reduce the number of features in the set
- Reinforcement learning
    - RL agent learning to perform a task by earning rewards or penalties based on performance
    - Example: a robot vacuum learns to navigate a room by encouraging movement that brings it over new locations and penalizes movement that runs into objects or goes over already cleaned areas
