# SymPy Mini Teaching Module (Beginner)

**Audience:** Students who know basic Python + algebra

**Learning goals:**
- Create symbols and symbolic expressions
- Simplify/expand/factor
- Solve equations
- Differentiate/integrate
- Work with matrices
- Convert symbolic expressions to numerical functions (`lambdify`)

---
## 0. Setup

```python
import sympy as sp
sp.__version__
```

### Pretty printing (optional)
In Jupyter, SymPy can display nicer math output.

```python
sp.init_printing(use_unicode=True)
```

---
## 1. Symbols and Expressions

```python
x, y = sp.symbols('x y')
expr = (x + 2*y)**2
expr
```

### Expand / Factor

```python
sp.expand(expr)
```

```python
sp.factor(sp.expand(expr))
```

### Simplify

```python
expr2 = (x**2 - y**2) / (x - y)
expr2
```

```python
sp.simplify(expr2)
```

---
## 2. Substitution and Evaluation
Use `.subs` to substitute values or other expressions.

```python
expr3 = sp.sin(x) + x**2
expr3.subs(x, sp.pi/2)
```

```python
expr3.subs(x, 3).evalf()  # numeric approximation
```

---
## 3. Solving Equations
Solve for unknowns with `sp.solve`.

### One equation

```python
sp.solve(sp.Eq(x**2 - 5*x + 6, 0), x)
```

### Two equations (system)

```python
sol = sp.solve([sp.Eq(x + y, 5), sp.Eq(x - y, 1)], [x, y])
sol
```

---
## 4. Calculus: Derivatives and Integrals

### Derivative

```python
f = sp.sin(x) * sp.exp(x)
sp.diff(f, x)
```

### Second derivative

```python
sp.diff(f, x, 2)
```

### Indefinite integral

```python
sp.integrate(sp.cos(x), x)
```

### Definite integral

```python
sp.integrate(sp.sin(x), (x, 0, sp.pi))
```

### Limits

```python
sp.limit(sp.sin(x)/x, x, 0)
```

---
## 5. Series Expansion (Optional but Useful)

```python
sp.series(sp.exp(x), x, 0, 6)  # up to x^5
```

---
## 6. Matrices

```python
A = sp.Matrix([[1, 2],
               [3, 4]])
A
```

```python
A.det()
```

```python
A.inv()
```

```python
A.eigenvals()  # eigenvalues with multiplicities
```

---
## 7. Convert Symbolic → Numeric with lambdify
This is useful when you want symbolic math but also fast numerical evaluation.

```python
g = sp.sin(x) + x**2
g_numeric = sp.lambdify(x, g, 'numpy')

import numpy as np
xs = np.linspace(-2, 2, 5)
g_numeric(xs)
```
