- These data points seem to be able to be fitted by 

$$y = ax^{-bx} + c$$

- Use `curve_fit` in `scipy` to find the coefficients $a$, $b$, and $c$.
- Use an intial guess of `p0=[15, 0.001, 5]`, however, you may see a non-convergency. You may need to set bounds of these coefficients.


```python
# put your code here

import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# exponential model
def exp_model(x, a, b, c):
    return a*np.exp(-b*x) + c


X = Data["tensile strength"].values
Y = Data["elongation"].values

# initial guess
p0 = [15, 0.001, 5]

# fit

popt, pcov = curve_fit(
    exp_model,
    X,
    Y,
    p0=[15, 0.001, 5],
    bounds=([0, 0, 0], [100, 0.01, 40])
)

# fitted parameters
a_fit, b_fit, c_fit = popt

print("a =", a_fit)
print("b =", b_fit)
print("c =", c_fit)
```

- Plot the obtained function with the original data points. Use different colors for the curve and points.

```python
# put your code here

# smooth fitted curve
x_fit = np.linspace(1000, 2500, 200)
y_fit = exp_model(x_fit, *popt)

# plot
plt.figure(figsize=(5,4))

plt.scatter(X, Y, s=80, label='data')
plt.plot(x_fit, y_fit, 'r', linewidth=2, label='exponential fit')

plt.xlabel("tempeature")
plt.ylabel("Young's modulus")
plt.grid(True)
plt.legend()

plt.show()
```

Build a SVM classifier to classify the grain sizes in images.

- Run the cells below to load in the image data and the corresponding labels.

- Browse the images. There are a total of 300 images. Each is of 256x256 pixels. You should also see label = 0, 1, 2, which correspond to fine, medium, coarse, respectively.

```python
from pathlib import Path
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt


base_dir = Path("grain_dataset")

images = []
labels = []

label_map = {
    "Fine":0,
    "Medium":1,
    "Coarse":2}

for class_name in label_map:

    folder = base_dir / class_name

    for fname in folder.glob("*.png"):

        img = Image.open(fname).convert("L")
        
        img = np.array(img)
        
        images.append(img)

        labels.append(label_map[class_name])

images = np.array(images)

labels = np.array(labels)

print(images.shape)
print(labels.shape)
```

```python
#Cool slider to browse all of the images.
from ipywidgets import interact
def browse_images(images, labels):
    n = len(images)
    def view_image(i):
        plt.imshow(images[i], cmap='viridis', interpolation='nearest')
        plt.title('%s' % f"label={labels[i]}")
        plt.axis('off')
        plt.show()
    interact(view_image, i=(0,n-1))
```

```python
browse_images(images, labels)
```

- To use the images in the training data. We need to `flatten` each image, i.e., convert a 256x256 image to a 1x65536 vector.
- Create a feature data from the images to a 300x65536 array.

```python
X = images.reshape(images.shape[0], -1)
y = labels

print("X shape:", X.shape)
print("y shape:", y.shape)
```

- Build a SVM classifier to classify the images.
- Set train-test split to be 0.2
- Print accuracy and classification_report
- Plot confusion matrix.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay, classification_report

import matplotlib.pyplot as plt

# normalize pixel values
X = X / 255.0

# train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.25,
    random_state=0,
    stratify=y)

# scale features
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# SVM classifier
svm_model = SVC(
    kernel="rbf",
    C=1.0,
    gamma="scale")

svm_model.fit(X_train_scaled, y_train)

# prediction
y_pred = svm_model.predict(X_test_scaled)

# accuracy
print("Accuracy:", accuracy_score(y_test, y_pred))

# classification report
print(classification_report(
    y_test,
    y_pred,
    target_names=["Fine", "Medium", "Coarse"]))

# confusion matrix
cm = confusion_matrix(y_test, y_pred)

disp = ConfusionMatrixDisplay(
    confusion_matrix=cm,
    display_labels=["Fine", "Medium", "Coarse"]
)

disp.plot()
plt.title("SVM Classification of Synthetic Grain Images")
plt.show()
```
