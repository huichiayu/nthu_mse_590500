## The Universal Approximation Theorem

As a group, please discuss how (if at all) the theorem below relates to the topic at hand.


> In the mathematical theory of artificial neural networks, the universal approximation theorem states that a feed-forward network with a single hidden layer containing a finite number of neurons (i.e., a multilayer perceptron), can approximate continuous functions on compact subsets of $R^n$, under mild assumptions on the activation function. The theorem thus states that simple neural networks can represent a wide variety of interesting functions when given appropriate parameters; however, it does not touch upon the algorithmic learnability of those parameters.
>
>One of the first versions of the theorem was proved by George Cybenko in 1989 for sigmoid activation functions.
>
>Kurt Hornik showed in 1991 that it is not the specific choice of the activation function, but rather the multilayer feedforward architecture itself which gives neural networks the potential of being universal approximators.
<p style="text-align: right;">From: Wikipidia - https://en.wikipedia.org/wiki/Universal_approximation_theorem</p>

### Some Math
Let $\varphi(\cdot)$ be a nonconstant, bounded, and monotonically-increasing continuous function. Let $I_m$ denote the $m$-dimensional unit hypercube $[0,1]^m$. The space of continuous functions on $I_m$ is denoted by $C(I_m)$. Then, given any function $f\in C(I_m)$ and $\epsilon>0$, there exists an integer $N$, real constants $v_i, b_i \in \mathbb{R}$ and real vectors $\mathbf{w}_i \in \mathbb{R}^m$, where $i = 1, \ldots, N$ such that if we define: 
$$F(\mathbf{x}) = \sum_{i=1}^N v_i \cdot\varphi \big(\langle\mathbf{w}_i , \mathbf{x}\rangle + b_i\big) $$
then 
$$|F(\mathbf{x}) - f(\mathbf{x})| < \epsilon $$

for all $x\in I_m$. In other words, functions of the form $F(\mathbf{x})$ are dense in $C(I_m)$.

## Neural network explained.

Here is an arbitrary senario. We have 3 students. Their **hours of sleeping** and **hours of studying** are 3, 5 ,10 and 5, 1, 2, respectively, before an exam. Their scores are 75, 82, 93, respectively. We will build a neural network model to predict a student's exam score based on his/her hours of sleeping and of studying.

In this case, we have 3 data points (each represents one student). Each data point contains 2 feature values and 1 output value.

**Features:** hours of sleeping and hours of studying.

**Output value:** exam score.

# 1. Build input vector and output vector

First, we will build the forward propgation path.  We will set up a simple neural network. Let's use 1 input layer, 1 hidden layer, and 1 output layer. Since we have 2 input feature values, the input layer will have 2 neurons. Similarly, we will have only 1 output value, thus the output layer has only 1 neuron. Let's assign 3 neurons to the hidden layer. See the figure below.

$$~$$

* We assign 2 weights to each neuron in the hidden layer. For example, $w_{11}^{(h)}$ and $w_{21}^{(h)}$ are the weight for $x_1$ and $x_2$ to neuron 1. The $j$-th neuron contains $w_{ij}^{(h)}$ weights.

* Since we have 3 neurons in the hidden layer, we need to assign 3 weight between the hidden and output layer. For example, $w_{11}^{(o)}$ is the weight for neuron 1 in the hidden layer to the neuron on the output layer. The values propagate from the hidden to the ourput laye are $a_1 w_{11}^{(o)} + a_2 w_{21}^{(o)} + a_3 w_{31}^{(o)} = \zeta$, i.e., $\zeta = \sum_{i} a_i w_{i1}^{(o)}$$.

Each neuron contains an activation function.

For one data point, the **forward path** follows

1. Multiple the two feature values with the hidden layer weights and sum them up. For example, at neurom 1 in the hidden layer: $z_1 = x_1 w_{11}^{(h)} + x_2 w_{21}^{(h)}$, i.e., $z_j = \sum x_i w_{ij}^{(h)}$.
2. Substitute the value $z_i$ value to the activation function: $a_i = f(z_i)$.
3. Multiple the $a$ values with the weights to the output layer and sum them up: $\zeta = a_1 w_{11}^{(o)} + a_2 w_{21}^{(o)} + a_3 w_{31}^{(o)}$.
4. Substitute $\zeta$ to the activation function: $\hat{y} = f(\zeta)$. Then, $\hat{y}$ is the predicted output.

Next, we evaluate the cost function, i.e., the error between predicted output and the true value of the score. Here, we use square error to quantify the error.

5. $J = \frac{1}{2}(y - \hat{y})^2$. This cost function $J$ is a function of the two sets of variables ($w_{ij}^{(h)}$ and $w_{ij}^{(o)}$).

**Backward correction**: We should realize that the purpose of training models is to **find a set of weights that minimize the cost function**, the smaller cost function, the higher accuracy the model is. The minimum of $J$ is where $\partial J/\partial w = 0$. 



>**Gradient descent method**
>
>To find the minimum of a function $f(x)$, we can use gradient descent method. We know that $-f'(x)$ is pointing to the direction of minimum point of $f(x)$. So, if we are at the position $x_n$, we can move by a small step toward the minimum direction by $x_{n+1} = x_{n} -f'(x_{n})\eta$, where $\eta$ is a small step size. By repeating this process, we can gradually and evantually reach the minimum point of $f(x)$. See the cell below for an illustration.


```python
import numpy as np

# function for plotting
import matplotlib.pylab as plt
from IPython.display import display, clear_output
import time

def show_animation(delay=0.05):
    fig = plt.gcf()
    time.sleep(delay)       # Sleep for half a second to slow down the animation
    clear_output(wait=True) # Clear output for dynamic display
    display(fig)            # Reset display
    fig.clear()             # Prevent overlapping and layered plots

# Choose the x points
x = np.linspace(-20,20,401)

# Define the function (parabola)
def f(x):
    return 3*x**2 - 10*x + 4

#Derivative of F1 (also called the gradient)
def grad_f(x):
    return 6*x - 10


##Gradient Decent
x0 = -7                     # starting point for gradient descent
eta = 0.05                  # step size (a.k.a learning rate)
p = .0001                   # chosen precision
maxIter = 30                # maximum number of iterations

former_min = x0
iterative_mins = [former_min]
iterCounter = 1

while True:
    new_min = former_min - eta * grad_f(former_min)
    
    iterative_mins.append(new_min)
    if abs(former_min - new_min) <= p:
        print('Local min of function is %f' %new_min)
        print('Number of iterations: %d' %iterCounter)
        break
    else:
        former_min = new_min
        
    if iterCounter == maxIter:
        print('Local min not reached')
        break
    else:
        iterCounter += 1       

    plt.figure(0)
    plt.plot(x, f(x), lw=3)
    plt.ylim(-10,100)
    plt.xlim(-10,10)
    plt.grid(True)
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.title('Iterative descent to x for f(x) minimum:'+str(new_min))
    plt.plot(iterative_mins, f(np.array(iterative_mins)), marker='o')
    show_animation(0.5)
```

>For multiple dimensions, the minimum of $J$ with variables ($w^{(h)}$ and $w^{(o)}$) can be reached following a similar way:
>$$\begin{bmatrix} w^{(o)} \\ w^{(h)}  \end{bmatrix}_{n+1} = \begin{bmatrix} w^{(o)} \\w^{(h)} \end{bmatrix}_{n} - \eta \begin{bmatrix} \frac{\partial J}{\partial w^{(o)}} \\ \frac{\partial J}{\partial w^{(h)}} \end{bmatrix}_{n}.$$ 

So, we need to calculate $\frac{\partial J}{\partial w^{(o)}}$ and $\frac{\partial J}{\partial w^{(h)}}$.

6. Let's work on $w_{i1}^{(o)}$ first:

$$\begin{split}\frac{\partial J}{\partial w_{i1}^{(o)}} & = \frac{\partial}{\partial w_{i1}^{(o)}} \bigg[\frac{1}{2}(y - \hat{y})^2 \bigg] =  - (y - \hat{y}) \frac{\partial \hat{y}}{\partial w_{i1}^{(o)}} = - (y - \hat{y}) \frac{\partial \hat{y}}{\partial \zeta} \frac{\partial \zeta}{\partial w_{i1}} = \delta_o a_i, \end{split}$$

where

$$\frac{\partial \hat{y}}{\partial \zeta} = f'(\zeta), ~~\frac{\partial \zeta}{\partial w_{i1}^{(o)}} = a_i,~~\text{and}~~\delta_o = -(y-\hat{y}) f'(\zeta).$$ 

Note that $y$ is a given constant. Its derivative is zero. Thus, only $\partial \hat{y}/\partial w_{i1}^{(o)}$ term remains in the equation above.


7. Now, let's work on $w_{ij}^{(h)}$.

$$\frac{\partial J}{\partial w_{ij}^{(h)}} = \frac{\partial}{\partial w_{ij}^{(h)}}\bigg[ \frac{1}{2}(y-\hat{y})^2 \bigg] = -(y -\hat{y} ) \frac{\partial \hat{y}}{\partial w_{ij}^{(h)}} = -(y -\hat{y} ) \frac{\partial \hat{y}}{\partial \zeta} \frac{\partial \zeta}{\partial a_j} \frac{\partial a_j}{\partial z_j}  \frac{\partial z_j}{\partial w_{ij}^{(h)} },$$

where $\partial \zeta / \partial a_i = w_{i1}^{(o)}$, $\partial a_j  / \partial z_j = f'(z_j)$, and $\partial z_j / \partial w_{ij}^{(h)} = x_i$. Thus,

$$\frac{\partial J}{\partial w_{ij}^{(h)}} = -(y -\hat{y} ) f'(\zeta) w_{i1}^{(o)}f'(z_j) x_i = \delta_o w_{i1}^{(o)}f'(z_j) x_i = \delta_h x_i. $$

**If we repeat the same process above, when submitting 1 data point into the forward propagation and backward correction, the weight will be adjusted toward the values that minimize the cost function $J$.** This is called training of a neural network.

## Matrix operation of the neuron network.

The above text explains the forward propgation and backward correction of 1 data point. In fact, it is much more convenient to work on a matrix containing all data points. It is called batch processing.

* For the input: it has 3 data points --> 3 rows. Each point has 2 features --> 2 column. Thus, the input vector will be a 3x2 array.

### Step 2:  Initialize random weights
Since we have 3 neurons in the hidden layer and there are 2 features. The weights in the hidden layer will be a 2x3 matrix. Each column corresponds to a neuron in the hidden layer.  The output layer weight will be a 3x1 column vector, each of which associates with a hidden-layer neuron. We can start with random values of these weights

```python
wh = np.random.randn(inputLayerSize, hiddenLayerSize)
wo = np.random.randn(hiddenLayerSize, outputLayerSize)

print(wh)
print(wo)
```

### Step 3: Multipuly the normalized input matrix by $w^{(h)}$
$$z = x  w^{(h)} $$ 
Here is the code using the numpy dot matrix. If you get an error you may have initilized the size of your variables incorrectly. Make sure the second dimention of `X` matches the first dimention of `wh`:

```python
z = np.matmul(X, wh)
print(z)
```

### Step 4: Implement and test the sigmoid function 

$$a(z) = \frac{1}{1 + e^{-z}} $$ 

The implemented  sigmoid function should take as input a numpy array and return a numpy array of the same dimension, with the function $f$ applied to each entry.

```python
# Apply sigmoid activation function to scalar, vector, or matrix
def sigmoid(z):
    return  1/(1+np.exp(-z)) 
```

### Step 5: Apply the sigmodal funciton to  $Z^{(2)}$

$$a = f(z)$$ 

Here is the code to apply the sigmod function to $z$ and display the results

```python
a = sigmoid(z)
a
```

###  Step 6:  multipuly $a$ by $w^{(o)}$ to get $\zeta$

$$\zeta = \mathbf{a} \mathbf{w}^{(o)} $$ 

Note that here $a$ is a matrix and $w^{(o)}$ is a column vector.

```python
zeta = np.matmul(a, wo)
print(zeta)
```

### Step 7: Apply the sigmod function again to $\zeta$ to produce $\hat{y}$

$$\hat{y} = f(\zeta)$$ 

```python
yHat = sigmoid(zeta)
yHat
```

### Step 8: Comparison
Now compare the estimation output ($\hat{y}$) to the actual output `y`. Calculate the cost function using mean square error.

$$J = \frac{1}{2} \sum_{i} \big(y-\hat{y}\big)^2$$

```python
print(y)
print(yHat)

J = 0.5*np.sum(y-yHat,axis=0)
print(J)
```

### Step 9: Derivative of sigmoid function 

The derivative of the activation function is 

$$f'(z) = \frac{e^{-z}}{(1+e^{-z})^2}$$

```python
def My_sigmoidPrime(s):
    #derivative of sigmoid
    return np.exp(-s)/(1 + np.exp(-s))**2
```

### Step 10: Backward correction.

Following the gradient descent method for multiple variables, we need to calculate the derivatives of the cost function with respect to the weights.

$$\nabla J = \begin{bmatrix} \frac{\partial J}{\partial w^{(o)}} \\ \frac{\partial J}{\partial w^{(h)}} \end{bmatrix}$$

Let's calculate that for $w^{(o)}$.

$$\frac{\partial J}{\partial w^{(o)}}= -(y-\hat{y})f'(\zeta)a_i = \delta_o a_i.$$


```python
# backward propgate through the network
delta_o = np.multiply(-(y - yHat), My_sigmoidPrime(zeta))
dJdwo = np.dot(a.T, delta_o)
print(dJdwo)
```

$$\frac{\partial J}{\partial w^{(h)}}=\delta_o w_{i1}^{(o)} f'(z_j) x_i = \delta_h x_i.$$

```python
delta_h = np.dot(delta_o, wo.T)*My_sigmoidPrime(z)
dJdwh = np.dot(X.T, delta_h)

print(dJdwh)
```

### Step 11: Backward correction of weights

$$\begin{bmatrix} w^{(o)} \\w^{(h)}  \end{bmatrix}_{n+1} = \begin{bmatrix} w^{(o)} \\ w^{(h)}  \end{bmatrix}_{n} - \eta \begin{bmatrix} \frac{\partial J}{\partial w^{(o)}} \\  \frac{\partial J}{\partial w^{(h)}}\end{bmatrix}_{n}$$ 

```python
eta = 0.2

# adjusting second set (hidden --> output) weights
wo += -eta*dJdwo 

# adjusting first set (input --> hidden) weights
wh += -eta*dJdwh 

print(wo)
print(wh)
```

### The above code demonstrates a loop of forward propagation and backward correction. 

The cell below puts everything together to a code.

```python
## Neuron network class

class My_Neural_Network(object):
    def __init__(self):
    #parameters
        self.inputSize = 2
        self.outputSize = 1
        self.hiddenSize = 3

        #weights
        self.wh = np.random.randn(self.inputSize, self.hiddenSize) # (3x2) weight matrix from input to hidden layer
        self.wo = np.random.randn(self.hiddenSize, self.outputSize) # (3x1) weight matrix from hidden to output layer

    def My_forward(self, X):
        #forward propagation through our network
        self.z = np.dot(X, self.wh) # dot product of X (input) and first set of 3x2 weights
        self.a = self.My_sigmoid(self.z) # activation function
        self.zeta = np.dot(self.a, self.wo) # dot product of hidden layer (z2) and second set of 3x1 weights
        yHat = self.My_sigmoid(self.zeta) # final activation function
        return yHat 

    def My_sigmoid(self, s):
        # activation function 
        return 1/(1+np.exp(-s))

## ~~~~~~~~~~~~~~~~~~~~~~~~~~above this line is forward calculation~~~~~~~~~~~~     
    
    def My_sigmoidPrime(self, s):
        #derivative of sigmoid
        return np.exp(-s)/(1 + np.exp(-s))**2

    def My_backward(self, X, y, yHat):
        # backward propgate through the network
        delta_o = np.multiply(-(y - yHat), self.My_sigmoidPrime(self.zeta))
        dJdwo = np.dot(self.a.T, delta_o)
        
        delta_h = np.dot(delta_o, self.wo.T)*self.My_sigmoidPrime(self.z)
        dJdwh = np.dot(X.T, delta_h)

        # adjusting first set (input --> hidden) weights
        eta = 0.2
        self.wh += -eta*dJdwh # <== fill the blank
        
        # adjusting second set (hidden --> output) weights
        self.wo += -eta*dJdwo # <== fill the blank
        
## ~~~~~~~~~~~~~~~~~~~~~~~~~above this line is backward propagation~~~~~~~~~~~~        

    def My_trainer(self, X, y):
        yHat = self.My_forward(X)
        self.My_backward(X, y, yHat)

## ~~~~~~~~~~~~~~~~~~~~~~~~~above this line is training~~~~~~~~~~~~~~~~~~~~~~~~

    def My_weights(self):
        return self.wh, self.wo 
```

In the cell below, we initiate a My_Neural_Network call with the name of MNN. Train it 10000 times.

```python
MNN = My_Neural_Network()

LSi = []
for i in range(10000): # trains the NN 10,000 times    
    MNN.My_trainer(X, y)   
    
    LSi.append(np.mean(np.square(y - MNN.My_forward(X))))
    

wh, wo = MNN.My_weights()

print("Input: \n" + str(X) ) 
print("Actual Output: \n" + str(y) ) 
print("Predicted Output: \n" + str(MNN.My_forward(X)) ) 
print("MSE: \n" + str(LSi[-1]) )   # mean sum squared loss
print("\n")

print(wh)
print(wo)
```

```python
plt.plot(LSi, linewidth= 2)
plt.grid(True)
plt.xlabel("epoch")
plt.ylabel("MSE")
plt.title("MSE over training")
plt.show()
```
