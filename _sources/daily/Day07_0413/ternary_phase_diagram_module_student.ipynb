## 1. Background

For a ternary system with components $A$, $B$, and $C$, the composition is often written as:

$$x_A + x_B + x_C = 1.$$

Because the three mole fractions must sum to 1, only **two** of them are independent.  
That is why a ternary composition can be shown on a **2D triangular plot** rather than in 3D.

A ternary phase diagram is useful for:

- alloy design
- ceramic composition maps
- battery materials
- slags, glasses, and oxides
- polymer blends
- any 3-component composition space

## 2. Geometry of the Gibbs triangle

A convenient way to display ternary compositions is to place the three pure components at the triangle corners:

- top corner: pure $A$
- left corner: pure $B$
- right corner: pure $C$

A point inside the triangle corresponds to a mixture of the three.

### Cartesian mapping used in this notebook

We will use the following triangle coordinates:

- $A = (0.5, \sqrt{3}/2)$
- $B = (0, 0)$
- $C = (1, 0)$

If a composition is $(x_A, x_B, x_C)$ and $x_A + x_B + x_C = 1$, then the Cartesian location is:

$$\mathbf{r} = x_A \mathbf{r}_A + x_B \mathbf{r}_B + x_C \mathbf{r}_C.$$

This is a barycentric-coordinate interpretation. Below is a code snippet to convert Cartesian coordinates to ternary coordinates (Gibbs triangle).

```python
def ternary_to_cartesian(xA, xB, xC):
    """
    Convert ternary composition to 2D Cartesian coordinates
    for an equilateral Gibbs triangle.
    Assumes xA + xB + xC = 1.
    """
    x = 0.5 * xA + xC
    y = (np.sqrt(3) / 2.0) * xA
    return x, y

# Test the corners
corners = {
    "A": (1.0, 0.0, 0.0),
    "B": (0.0, 1.0, 0.0),
    "C": (0.0, 0.0, 1.0),
}

for name, comp in corners.items():
    print(name, comp, "->", ternary_to_cartesian(*comp))
```

```python
# Plot the Gibbs triangle in standard Cartesian axes 
fig, ax = plt.subplots(figsize=(5, 5))

# Triangle edges
triangle = np.array([
    ternary_to_cartesian(1, 0, 0),  # A
    ternary_to_cartesian(0, 1, 0),  # B
    ternary_to_cartesian(0, 0, 1),  # C
    ternary_to_cartesian(1, 0, 0)   # close
])

# plot line segments
ax.plot(triangle[:, 0], triangle[:, 1], linewidth=2)

# Labels
A = ternary_to_cartesian(1, 0, 0)
B = ternary_to_cartesian(0, 1, 0)
C = ternary_to_cartesian(0, 0, 1)

ax.text(A[0], A[1] + 0.04, "A", ha="center", fontsize=14)
ax.text(B[0] - 0.04, B[1] - 0.03, "B", ha="center", fontsize=14)
ax.text(C[0] + 0.04, C[1] - 0.03, "C", ha="center", fontsize=14)

ax.set_aspect("equal")
ax.set_xlim(-0.1, 1.1)
ax.set_ylim(-0.1, 1.0)
ax.set_title("Gibbs Triangle in Cartesian Coordinates")
ax.axis("off")
plt.show()
```

## 5. Building a composition grid

To make a ternary contour map, we need many points inside the triangle.

A simple way is to generate all compositions on a discrete grid such that:

$$x_A + x_B + x_C = 1$$

Note that **every valid composition must satisfy the constraint**.

```python
def make_ternary_grid(n=40):
    """
    Return arrays of ternary compositions on a uniform grid.
    n controls the resolution.
    """
    A_list, B_list, C_list = [], [], []
    for i in range(n + 1):
        for j in range(n + 1 - i):
            k = n - i - j
            A_list.append(i / n)
            B_list.append(j / n)
            C_list.append(k / n)
    return np.array(A_list), np.array(B_list), np.array(C_list)

A_grid, B_grid, C_grid = make_ternary_grid(n=40)
print("Number of grid points:", len(A_grid))
print("Check composition sum range:", np.min(A_grid + B_grid + C_grid), np.max(A_grid + B_grid + C_grid))
```

## 6. A synthetic property map

We often want a smooth scalar field over the ternary diagram.  
This could represent:

- density
- conductivity
- elastic modulus
- melting temperature
- free energy (qualitatively)

Below, we define a synthetic property function with a few peaks and valleys.

```python
def synthetic_property(xA, xB, xC):
    """
    A smooth illustrative property field over composition space.
    This is not tied to a real material system.
    """
    term1 = 1.2 * np.exp(-((xA - 0.65)**2 + (xB - 0.20)**2 + (xC - 0.15)**2) / 0.02)
    term2 = 0.8 * np.exp(-((xA - 0.20)**2 + (xB - 0.55)**2 + (xC - 0.25)**2) / 0.03)
    term3 = 0.6 * np.exp(-((xA - 0.15)**2 + (xB - 0.15)**2 + (xC - 0.70)**2) / 0.025)
    background = 0.25 * (xA*xB + xB*xC + xC*xA)
    return term1 + term2 + term3 + background

Z_prop = synthetic_property(A_grid, B_grid, C_grid)
Z_prop.min(), Z_prop.max()
```

```python
fig = plt.figure(figsize=(5, 5))
ax = fig.add_subplot(projection="ternary")

tcf = ax.tricontourf(A_grid, B_grid, C_grid, Z_prop, levels=20)
ax.tricontour(A_grid, B_grid, C_grid, Z_prop, levels=10, linewidths=0.5)

ax.set_tlabel("A")
ax.set_llabel("B")
ax.set_rlabel("C")
ax.set_title("Synthetic Property Map on a Ternary Diagram")
ax.grid(True)

cbar = plt.colorbar(tcf, ax=ax, shrink=0.85)
cbar.set_label("Property value")

plt.show()
```

## 7. A simple free-energy-like function

Now let us build a **toy free energy** to illustrate ideas such as:

- stable composition regions
- local minima
- phase separation tendencies

A common teaching form is:

$$
G = \sum_i x_i \ln x_i + \Omega_{AB} x_A x_B + \Omega_{BC} x_B x_C + \Omega_{CA} x_C x_A
$$

The first term resembles an ideal mixing term.  
The interaction parameters $\Omega_{ij}$ can raise or lower mixing tendency.

Below we use a simple illustrative model, not a calibrated thermodynamic database.

```python
def safe_xlogx(x):
    x = np.clip(x, 1e-12, None)
    return x * np.log(x)

def toy_free_energy(xA, xB, xC, T=1.0, OmegaAB=2.8, OmegaBC=3.4, OmegaCA=2.2):
    ideal = T * (safe_xlogx(xA) + safe_xlogx(xB) + safe_xlogx(xC))
    excess = OmegaAB*xA*xB + OmegaBC*xB*xC + OmegaCA*xC*xA

    # Add a couple of weak wells just to make the pedagogical picture richer
    wells = (
        -0.18 * np.exp(-((xA - 0.70)**2 + (xB - 0.20)**2 + (xC - 0.10)**2) / 0.01)
        -0.12 * np.exp(-((xA - 0.18)**2 + (xB - 0.62)**2 + (xC - 0.20)**2) / 0.015)
    )
    return ideal + excess + wells

G = toy_free_energy(A_grid, B_grid, C_grid)
G.min(), G.max()
```

```python
fig = plt.figure(figsize=(5, 5))
ax = fig.add_subplot(projection="ternary")

tcf = ax.tricontourf(A_grid, B_grid, C_grid, G, levels=25)
ax.tricontour(A_grid, B_grid, C_grid, G, levels=12, linewidths=0.5)

ax.set_tlabel("A")
ax.set_llabel("B")
ax.set_rlabel("C")
ax.set_title("Toy Free-Energy Surface Projected onto a Ternary Diagram")
ax.grid(True)

cbar = plt.colorbar(tcf, ax=ax, shrink=0.85)
cbar.set_label("Free-energy-like value")

plt.show()
```

## 8. Highlighting low-energy regions

Low-energy regions are often associated with preferred compositions or stable phases.  
In real thermodynamics, the full story is more subtle because phase equilibria are determined by the **convex hull / common tangent construction**, not just by pointwise minima.

Still, it is very useful pedagogically to highlight low-$G$ regions first.

```python
threshold = np.percentile(G, 12)  # lowest 12% of points
mask = G <= threshold

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(projection="ternary")

ax.scatter(A_grid, B_grid, C_grid, s=8, alpha=0.15)
ax.scatter(A_grid[mask], B_grid[mask], C_grid[mask], s=18)

ax.set_tlabel("A")
ax.set_llabel("B")
ax.set_rlabel("C")
ax.set_title("Low Free-Energy Regions (Illustrative)")
ax.grid(True)

plt.show()
```

### 3D surface

Here, we can make a 3D interactive surface to examine the given free energy surface.

```python
from scipy.spatial import Delaunay
import plotly.graph_objects as go

X, Y = ternary_to_cartesian(A_grid, B_grid, C_grid)

tri = Delaunay(np.column_stack([X, Y]))
simplices2d = tri.simplices

i = simplices2d[:, 0]
j = simplices2d[:, 1]
k = simplices2d[:, 2]

fig = go.Figure()

fig.add_trace(go.Mesh3d(
    x=X,
    y=Y,
    z=G,
    i=i,
    j=j,
    k=k,
    intensity=G,
    opacity=0.99,
    showscale=True,
    colorbar=dict(title="G"),
    hovertemplate=(
        "A=%{customdata[0]:.3f}<br>"
        "B=%{customdata[1]:.3f}<br>"
        "C=%{customdata[2]:.3f}<br>"
        "G=%{z:.3f}<extra></extra>"
    ),
    customdata=np.column_stack([A_grid, B_grid, C_grid])
))

A = np.array([0.0, 0.0])
B = np.array([1.0, 0.0])
C = np.array([0.5, np.sqrt(3)/2])
boundary = np.array([A, B, C, A])

z0 = float(G.min()) - 0.15 * (float(G.max()) - float(G.min()))

fig.add_trace(go.Scatter3d(
    x=boundary[:, 0],
    y=boundary[:, 1],
    z=np.full(len(boundary), z0),
    mode="lines",
    name="Ternary boundary"
))

fig.add_trace(go.Scatter3d(
    x=[A[0], B[0], C[0]],
    y=[A[1], B[1], C[1]],
    z=[z0, z0, z0],
    mode="text",
    text=["A", "B", "C"],
    textposition="top center",
    name="Corner labels"
))

fig.update_layout(
    title="Interactive 3D Ternary Free-Energy Surface",
    scene=dict(
        xaxis_title="Triangle x",
        yaxis_title="Triangle y",
        zaxis_title="G",
        aspectmode="data",
        camera=dict(eye=dict(x=1.6, y=1.4, z=1.1))
    ),
    width=920,
    height=760
)

fig.show()
```

## 9. Tie-lines and tie-triangles

### Tie-line
In a two-phase region, two equilibrium phase compositions are connected by a **tie-line**.  
A bulk composition lying on that tie-line can decompose into those two phases.

### Tie-triangle
In a three-phase region, three equilibrium phase compositions define a **tie-triangle**.  
Any overall composition inside that triangle can be represented as a mixture of the three phases.

Below we draw **illustrative** tie-lines and a tie-triangle.  
These are schematic teaching graphics, not computed phase equilibria.

```python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(projection="ternary")

# Background contour
ax.tricontourf(A_grid, B_grid, C_grid, G, levels=25)

# Example two-phase tie-line endpoints
P_alpha = (0.72, 0.20, 0.08)
P_beta  = (0.18, 0.60, 0.22)

# Example three-phase triangle vertices
P1 = (0.75, 0.18, 0.07)
P2 = (0.20, 0.65, 0.15)
P3 = (0.12, 0.18, 0.70)

# Draw tie-line
ax.plot([P_alpha[0], P_beta[0]],
        [P_alpha[1], P_beta[1]],
        [P_alpha[2], P_beta[2]],
        linewidth=2)

ax.scatter(*P_alpha, s=60)
ax.scatter(*P_beta, s=60)
ax.text(*P_alpha, "  α")
ax.text(*P_beta,  "  β")

# Draw tie-triangle
triA = [P1[0], P2[0], P3[0], P1[0]]
triB = [P1[1], P2[1], P3[1], P1[1]]
triC = [P1[2], P2[2], P3[2], P1[2]]
ax.plot(triA, triB, triC, linewidth=2)

for label, P in zip(["γ1", "γ2", "γ3"], [P1, P2, P3]):
    ax.scatter(*P, s=60)
    ax.text(*P, "  " + label)

ax.set_tlabel("A")
ax.set_llabel("B")
ax.set_rlabel("C")
ax.set_title("Illustrative Tie-Line and Tie-Triangle")
ax.grid(True)

plt.show()
```

## 10. Example: lever-rule idea on a tie-line

Suppose a bulk composition lies on the tie-line between phases $\alpha$ and $\beta$.  
Then the overall composition can be written as:

$$
\mathbf{x}^{(0)} = f_\alpha \mathbf{x}^{(\alpha)} + f_\beta \mathbf{x}^{(\beta)}
$$

with

$$
f_\alpha + f_\beta = 1
$$

The fractions are determined by the lever rule.

Below is an illustrative interpolation along the tie-line.

```python
P_alpha = np.array([0.72, 0.20, 0.08])
P_beta  = np.array([0.18, 0.60, 0.22])

f_alpha = 0.35
f_beta = 1.0 - f_alpha

P_bulk = f_alpha * P_alpha + f_beta * P_beta
print("Bulk composition =", P_bulk)
print("Check sum =", P_bulk.sum())

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(projection="ternary")

ax.plot([P_alpha[0], P_beta[0]],
        [P_alpha[1], P_beta[1]],
        [P_alpha[2], P_beta[2]],
        linewidth=2)

ax.scatter(*P_alpha, s=80)
ax.scatter(*P_beta, s=80)
ax.scatter(*P_bulk, s=80)

ax.text(*P_alpha, "  α")
ax.text(*P_beta, "  β")
ax.text(*P_bulk, "  bulk")

ax.set_tlabel("A")
ax.set_llabel("B")
ax.set_rlabel("C")
ax.set_title("Lever-Rule Concept on a Tie-Line")
ax.grid(True)

plt.show()
```
