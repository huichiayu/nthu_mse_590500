## 1. What is the Shell?
The shell is a program that lets you interact with the operating system using text commands.

Check your shell:
```bash
echo $SHELL
```

## 2. Basic Navigation

```bash
pwd
ls
ls -l
ls -a
cd folder_name
cd ..
cd ~
```

## 3. Creating and Managing Files

```bash
mkdir my_folder
touch file.txt
cp file.txt copy.txt
mv copy.txt newname.txt
rm newname.txt
rm -r my_folder
```

## 4. Viewing File Contents

```bash
cat file.txt
less file.txt
head file.txt
tail file.txt
```

## 5. Redirection and Pipes

```bash
echo "Hello" > hello.txt
echo "World" >> hello.txt
ls -l | less
cat file.txt | grep phase transformation
```

## 6. Wildcards

```bash
ls *.txt
ls data_?.csv
```

## 7. Permissions

```bash
ls -l
chmod u+x script.sh
```

## 8. Useful Shortcuts

- Up arrow: previous command
- `history`
- `Ctrl + R`
- `Tab` autocomplete
