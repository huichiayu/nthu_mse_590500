## 0. Setup

```python
import numpy as np
import matplotlib.pyplot as plt
```

---
## 1. Your First Plot
Plot a simple sine function.

```python
x = np.linspace(0, 2*np.pi, 100)
# x

y = np.sin(x)
# y
plt.figure()
plt.plot(x, y)
plt.show()
```

### Add labels and title

```python
plt.figure()
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Sine Function')
plt.show()
```

---
## 2. Understanding Figure and Axes
Matplotlib uses a figure containing axes objects.

```python
fig, ax = plt.subplots()
ax.plot(x, np.cos(x))
ax.set_xlabel('x')
ax.set_ylabel('cos(x)')
ax.set_title('Cosine Function')
plt.show()
```

---
## 3. Multiple Curves on One Plot

```python
plt.figure()
plt.plot(x, np.sin(x), 'rd:')
plt.plot(x, np.cos(x), 'bo--')
plt.legend(['sin(x)', 'cos(x)'])
plt.title('Sine and Cosine')
plt.show()
```

---
## 4. Scatter Plot

```python
np.random.seed(0)
x = np.random.rand(50)
y = np.random.rand(50)

plt.figure()
plt.scatter(x, y)
plt.title('Random Scatter')
plt.show()
```

---
## 5. Histogram

```python
data = np.random.randn(1000)

plt.figure()
plt.hist(data, bins=30)
plt.title('Histogram')
plt.show()
```

---
## 6. Bar Plot

```python
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 8]

plt.figure()
plt.bar(categories, values)
plt.title('Bar Chart')
plt.show()
```

---
## 7. Adjusting Limits and Grid

```python
plt.figure()
plt.plot(x, np.sin(x))
plt.xlim(0, np.pi)
plt.ylim(-1, 2)
plt.grid(True)
plt.title('Sine (Zoomed)')
plt.show()
```

---
## 8. Saving Figures

```python
plt.figure()
plt.plot(x, np.sin(x))
plt.title('Saved Figure Example')
plt.savefig('example_plot.png')
plt.show()
```
