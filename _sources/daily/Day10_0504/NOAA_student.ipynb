### Use FFT to denoise data. 

Calculate the Fast Fourier Transform of the (NOAA) surface temperature data and display it in the frequency domain. Remember that the frequency domain is symmetric so go ahead and just display the first half. 

- Let's just work on the 2nd half of the temperature data.
- Use FFT to find the amplitudes associated with the frequencies.

```python
import numpy as np

pp = len(air_temperatures)
temps = air_temperatures[int(pp/2):]
myV = np.fft.fft(temps)

myV_half = np.abs(myV[:int(len(myV)/2)])

plt.plot(abs(myV_half))
```

- Write some code to find the five frequencies with largest amplitudes in the frequency space (make sure you ignore symmetry).
