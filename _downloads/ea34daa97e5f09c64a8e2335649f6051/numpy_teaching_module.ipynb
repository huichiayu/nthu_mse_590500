## 1. Why NumPy?
Python lists don’t do elementwise math by default. NumPy arrays do.

### Python list example

```python
a = [1, 2, 3]
b = [4, 5, 6]
a + b  # This concatenates lists, not elementwise addition
```

### NumPy array example

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
c = a + b  # Elementwise addition
c
```

## 4. Indexing and Slicing

```python
A = np.array([[10, 20, 30],
              [40, 50, 60]])
A
```

### Access elements

```python
A[0, 1]
```

### Row / column

```python
A[0, :2]  # first row
```

```python
A[:, 1]  # second column
```

### Submatrix

```python
A[:, 1:]  # all rows, columns 1 to end
```

### Views vs copies
Slicing usually returns a **view** (shares data). Mutating the slice can mutate the original.

```python
B = A[:, 1:]
B[0, 0] = 999
A  # notice A changed
```

If you want a copy, use `.copy()`:

```python
A = np.array([[10, 20, 30],
              [40, 50, 60]])
C = A[:, 1:].copy()
C[0, 0] = -1
print('C =\n', C)
print('A =\n', A)
```

## 6. Broadcasting
Broadcasting lets NumPy combine arrays of different shapes (when compatible).

```python
A = np.array([[1, 2, 3],
              [4, 5, 6]])
b = np.array([10, 20, 30])
A + b  # b is broadcast across rows
```

**Rule of thumb (simplified):** compare shapes from right to left; dimensions must match or one must be 1.

## 8. Matrix Operations
Be careful: `*` is elementwise, while `@` is matrix multiplication.

```python
A = np.array([[1, 2],
              [3, 4]])
B = np.array([[5, 6],
              [7, 8]])
A, B
```

### Elementwise multiplication

```python
A * B
```

### Matrix multiplication

```python
A @ B
```

```python
np.dot(A, B)
```

### Determinant / inverse (if invertible)

```python
np.linalg.det(A)
```

```python
h = np.linalg.inv(A)
h @ A
```

## 9. Performance Comparison
NumPy is fast because it uses optimized compiled code under the hood.

Run these a couple times to see typical behavior.

```python
import time

size = 1_000_000
python_list = list(range(size))
numpy_array = np.arange(size)
```

### Python loop sum

```python
start = time.time()
s = 0
for i in python_list:
    s += i
end = time.time()
print('Python time:', end - start)
print('Sum:', s)
```

### NumPy sum

```python
start = time.time()
s2 = np.sum(numpy_array)
end = time.time()
print('NumPy time:', end - start)
print('Sum:', s2)
```
