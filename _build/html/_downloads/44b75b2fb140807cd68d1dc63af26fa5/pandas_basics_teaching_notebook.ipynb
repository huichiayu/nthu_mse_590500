## 1. Import Pandas

Pandas is the standard Python library for working with tabular data.

```python
import pandas as pd

print(pd.__version__)
```

## 2. Create a Series

A `Series` is like a labeled one-dimensional array.

```python
temperatures = pd.Series([72, 75, 78, 74], index=['Mon', 'Tue', 'Wed', 'Thu'])
temperatures
```

## 3. Create a DataFrame

A `DataFrame` is a table with rows and columns.

```python
data = {
    'name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Ethan'],
    'age': [23, 25, 22, 24, 26],
    'major': ['Physics', 'Math', 'Biology', 'Chemistry', 'Math'],
    'score': [88, 92, 85, 91, 79]
}

df = pd.DataFrame(data)
df
```

## 4. Inspect the DataFrame

Useful methods for looking at your data quickly.

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

## 5. Select Columns

You can select one column or multiple columns.

```python
df['name']
```

```python
df[['name', 'score']]
```

## 6. Select Rows with `.iloc` and `.loc` 

- `.iloc` uses integer positions
- `.loc` uses labels

```python
df.iloc[0]
```

```python
df.iloc[1:3, ['name', 'major', 'score']]
```

```python
df.loc[2]
```

## 7. Filter Rows

Boolean conditions let you keep only rows that match a rule.

```python
df[df['score'] >= 90]
```

```python
df[df['major'] == 'Math']
```

```python
df[(df['score'] >= 85) & (df['age'] < 25)]
```

## 8. Add a New Column

You can create columns from existing columns.

```python
df['passed'] = df['score'] >= 85
df
```

```python
df['score_plus_5'] = df['score'] + 5
df
```

## 9. Basic Statistics

Pandas makes column-based calculations easy.

```python
df['score'].mean(), df['score'].max(), df['score'].min()
```

```python
df[['age', 'score']].mean()
```

## 10. Sort Values

Sorting is useful when ranking or organizing data.

```python
df.sort_values('score', ascending=False)
```

## 11. Group By

Group rows by a category and calculate summary values.

```python
df.groupby('passed')['score'].mean()
```

```python
df.groupby('passed').agg({
    'score': ['mean', 'max'],'age': 'mean'
})
```

## 12. Missing Values

Real datasets often contain missing data.

```python
df2 = df.copy()
df2.loc[2, 'score'] = None
df2.loc[4, 'major'] = None
df2
```

```python
df2.isna()
```

```python
df2.dropna()
```

```python
df2['score'] = df2['score'].fillna(df2['score'].mean())
df2['major'] = df2['major'].fillna('Unknown')
df2
```

## 13. Read and Write CSV Files

These are among the most common Pandas tasks.

```python
# Save a DataFrame to CSV
df.to_csv('students.csv', index=False)

# Read it back
students = pd.read_csv('students.csv')
students.head()
```
