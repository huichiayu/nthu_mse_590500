$$S = \int \frac{C_p}{T} dT$$.

1. Entropy of a substance at 298 K, $S_{298}$, will be a given value. In this example, $S_{PbO,s}(298) = 67.4$, $S_{Pb,s}(298) = 64.9$, and $S_{O_2,s}(298) = 205$ per mole.
2. Melting entropy is the melting enthalpy divided by temperature.
$$S_m = \frac{H_m}{T_m}$$.

**Start with solid Pb.**

```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np
```

```python
T = sp.symbols('T')

# solid Pb
Cp_Pbs = 23.6 + 9.75e-3 * T 

# entropy
S_Pbs = sp.integrate(Cp_Pbs/T, T)
S_Pbs
```

3. The entropy of Pb at 298 K is given as
$$S_{Pb,s} (T=298)=64.9$$
Thus, we need to normalize the value to be 64.9 at T = 298.

```python
# reference
Pbs_S_ref = S_Pbs.subs(T, 298)
Pbs_S_ref.evalf()

S_Pbs = S_Pbs - Pbs_S_ref + 64.9

# visualize
S_Pbs_plt = sp.lambdify(T, S_Pbs, "numpy")
Ts_vals = np.linspace(298, 600, 100)

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(Ts_vals, S_Pbs_plt(Ts_vals), 'r', label="Pbs")
ax.set_xlabel("T")
ax.set_ylabel("Entropy (J)")
ax.grid(True)
ax.legend()
```

**The metling entropy at T=600.** The entropy of liquid Pb at the melting is S_Pbs(600)+ S_m, which is the reference for calibrating the entropy of liquid Pb.
