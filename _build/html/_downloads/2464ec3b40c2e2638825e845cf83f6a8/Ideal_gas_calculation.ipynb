Ideal gas law:

$$PV = NRT$$

For an close system during isothermal process, $N$ and $T$ are fixed. Here we set $N = 1$ mol, $T=1000$ K, and $R=0.082$ lit$\cdot$atm/degree$\cdot$mol. Plot the P-V curve during a decompression from $P_1=20$ to $P_2 = 4$ atm. 

````python
import numpy as np
import matplotlib.pyplot as plt

# ideal gas constant
R = 0.082   # atm * liter / (K * mol)

# state 1
T1 = 1000   # K
P1 = 20     # atm

V1 = R * T1 / P1

# isothermal process
# state 2
P2 = 4      # atm
V2 = P1 * V1 / P2

print(V1, V2)

P = np.linspace(P1, P2, 40)

V = P1 * V1 / P



fig, ax = plt.subplots()

ax.plot(V, P, 'ko-', label='isothermal')
ax.set_xlabel('V (liter)')
ax.set_ylabel('P (atm)')
ax.grid(True)
````

For an adiabatic process, the P-V relationship is

$$\frac{P_2}{P_1} = \bigg(\frac{V_1}{V_2}\bigg)^\gamma,$$

$\gamma = 5/3$. Recall the derivation.

$$c_v dT = -P dV ~~\longrightarrow~~ c_v dT = -\frac{RT}{V} dV ~~\longrightarrow~~ c_v \frac{dT}{T} = -R\frac{dV}{V}$$

$$c_v \ln\bigg(\frac{T_2}{T_1}\bigg) = R \ln\bigg(\frac{V_1}{V_2}\bigg) ~~\longrightarrow~~ \bigg(\frac{T_2}{T_1}\bigg) = \ln\bigg(\frac{V_1}{V_2}\bigg)^\frac{R}{c_v} ~~\longrightarrow~~ \frac{P_2 V_2}{P_1 V_1} = \ln\bigg(\frac{V_1}{V_2}\bigg)^\frac{R}{c_v}.$$

````python
# adiabatic process
gamma = 5/3

V = V1 / (P/P1)**(1/gamma)

print(V[-1])

ax.plot(V, P, 'bo-', label='adiabatic')
ax.legend()

fig
````

Calculate the work done to the environment by the gas. 

$$W = \int_1^2 PdV.$$

Here we can use trapzoidal rule of numerical integration. Use `trapz()` in numpy library.

A small area between $x_i$ and $x_{i+1}$ can be approximated by

$$\text{Area}_i \approx \frac{1}{2}(y_i + y_{i+1})(x_{i+1}-x_i)$$

The total area is obtained by summing all these small areas:
$$
\int_a^b f(x)\,dx
\approx
\sum_{i=0}^{n-1}
\frac{1}{2}(y_i+y_{i+1})(x_{i+1}-x_i)
$$

An example code for trapzoidal method:
```bash
def trapz(x, y):

    n = len(x)
    area = 0.0

    for i in range(n-1):
        dx = x[i+1] - x[i]
        area += 0.5 * (y[i] + y[i+1]) * dx

    return area
```

Isothermal process

````python
import numpy as np

# parameters
P1 = 20
V1 = 4.1
V2 = 20.5

# pressure function for isothermal process
def P_iso(V):
    return P1 * V1 / V

# trapezoidal integration
n = 100
V = np.linspace(V1, V2, n+1)
Pvals = P_iso(V)

W = np.trapz(Pvals, V)

print("Work =", W)
````

Adiabatic process

````python
V2 = 10.768763998055444

n = 100
V = np.linspace(V1, V2, n+1)

def P_adi(V):
    return P1 * (V1/V)**gamma


Pvals = P_adi(V)

W = np.trapz(Pvals, V)

print("Work =", W)
````
